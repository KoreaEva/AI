## 텐서(Tensor)
- 일반적으로 텐서는 3차원 이상을 다룰 때 표현하는 방식이지만, 여기서는 어떠한 데이터를 표현할 때, 그 값 모두를 텐서라고 부르기로 함

      a = np.array([1, 2])
      b = np.array([[1, 2],
                    [3, 4]])
      c = np.array([10]) 
      d = np.array(3)  
  
    a, b, c, d 모두 텐서라고 지칭할 수 있음
  
- 랭크(rank): 텐서의 축을 나타내고, 넘파이(numpy)의 `ndim`으로 구할 수 있음

      a = np.array([1, 2])
        >> ndim = 1  
      b = np.array([[1, 2],
                    [3, 4]])
        >> ndim = 2
      d = np.array(3)
        >> ndim = 0
      
    위의 예시에서 알 수 있듯이 대괄호( [ ] )의 개수가 곧 랭크(축)의 값
    
      
- 크기(shape): 텐서의 각 축을 따라 얼마나 많은 차원이 있는지를 나타내며, 파이썬의 튜플(tuple) 형태

### 벡터의 합
- 같은 형상(shape)일 때, 각 원소별로 계산

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
c = a + b

print(c)
print(c.shape)
print(np.ndim(c))
```

### 벡터의 곱
-  $A = (x_1, \ x_2, \ x3, \ ..., \ x_n)$  
  $B = (y_1, \ y_2, \ y3, \ ..., \ y_n)$ 일 때,
- 원소곱
  - 같은 형상(shape)일 때, 각 원소별로 계산   
   
    $A \times B = (x_1, \ x_2, \ x_3, \ ..., \ x_n) \times (y_1, \ y_2, \ y_3, \ ..., \ y_n) \\   
    \qquad = (x_1y_1, \ x_2y_2, \ x_3y_3, \ ... \ ,  \ x_ny_n)  $
    
- 벡터곱(product, dot)
  - 두 1차원 벡터가 있을 때 **각각의 성분끼리의 곱을 모두 더하는 계산**  
  
    $A \bullet B \Rightarrow A \times B^T = (x_1, \ x_2, \ x_3, \ ..., \ x_n) \begin{pmatrix} y_1 \\ y_2 \\ y_3 \\ ... \\ y_n  \end{pmatrix} \\  
    \qquad = (x_1y_1 + \ x_2y_2 + \ x_3y_3 + \ ... + \ x_ny_n)  $
  

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
c = a * b

print(c)
print(c.shape)
print(np.ndim(c))
print('------------------------')

x = np.array([1,2,0])
y = np.array([0,2,1])
z = np.dot(x,y)

print(z)
print(z.shape)
print(np.ndim(z))
```

### 행렬 원소곱
- 같은 형상(shape)일 때 덧셈, 곱셈과 같은 연산은 원소별로 진행


```python
A = np.array([[1,2],[3,4]])
B = np.array([[10,10],[10,10]])

print('행렬 A\n', A)
print('행렬 B\n', B)
print('A * B\n', A*B)
```

### 행렬 점곱(내적, product)
- 1차원 벡터와 마찬가지로 앞 행렬의 열과 뒤 행렬의 행의 수가 같아야함

```python
M = np.array([[1,2],[3,4]])
N = np.array([[2,3,4],[2,3,4]])

print('행렬 M\n', M)
print('행렬 N\n', N)

L = np.dot(M,N)
print('행렬 L\n', L)
print(L.shape)
print(np.ndim(L))
```

```python
m = np.array([[1],[2],[3]])
n = np.array([[1],[2],[3]])
l = np.dot(m,n)

print(l)
```

### 역행렬
- 어떤 행렬 A가 있을 때, 곱해서 단위행렬(E)를 만드는 행렬 B가 존재한다면, 행렬 B는 A의 역행렬

```python
A = np.array([[1,2],[3,4]])
print(A)
B = np.linalg.inv(A)
print(B)
print(np.dot(A,B))
```

```python
B = np.array([[3,3],[2,2]])
print(np.linalg.inv(B))
```

### 전치행렬
- 행과 열을 바꾼 배열의 형태

![](https://upload.wikimedia.org/wikipedia/commons/e/e4/Matrix_transpose.gif)
<br /><sub>출처: https://ko.wikipedia.org/wiki/%EC%A0%84%EC%B9%98%ED%96%89%EB%A0%AC</sub>

```python
A = np.array([[1,2,3,],[4,5,6]])
print('A\n', A)
print('A.shape\n', A.shape)
print('----------------')
A_ = A.T
print('A의 전치행렬\n', A_)
print('(A.T).shape\n', A_.shape)
B = A_.T
print(B)
```

## 텐서 크기 변환
- reshape로 텐서의 크기 변환 가능
- **변환 전의 원소의 개수와 변환 이후의 텐서의 개수가 같아야** 함

```python
A = np.array([[1,2,3],[4,5,6]])
print('행렬 A\n', A)
print('A.shape:', A.shape)
print('--------------------')

B = A.reshape(6)
print('B\n', B)
print('B.shape:', B.shape)
```

```python
B = np.array([[[2,3,4],[5,6,7]],
             [[1,1,1],[1,1,1]]])

print('행렬 B\n', B)
print('B.shape:', B.shape)
print('-----------------')
B = B.reshape(3,4)
print('B\n', B)
print('B.shape:', B.shape)
```

- -1 을 통해 자동으로 형상을 지정 가능  
  원소의 갯수에 맞게 넘파이가 자동으로 형상을 지정  
      (2, 2, 3) -> (3, -1)     (O)
                -> (2, 1, 6)   (O)
                -> (2, -1, -1) (X)
                -> (2, 5, -1)  (X) 

```python
B = np.array([[[2,3,4],[5,6,7]],
             [[1,1,1],[1,1,1]]])

print('행렬 B\n', B)
print('B.shape:', B.shape)
print('-----------------')
B = B.reshape(4,-1)
print('B\n', B)
print('B.shape:', B.shape)
```
